## Section 7: Data Cleaning — Identifying and Fixing Issues
Code block leveraged and reused from: Seminar Session — Data Cleaning Lab

```python
# ISSUE 1: age column — missing values and outlier (age = 123)
print('Missing age values:', df_clean['age'].isnull().sum())
print('Age outliers (>100):', (df_clean['age'] > 100).sum())
print('Age max value:', df_clean['age'].max())
```

```python
# FIX ISSUE 1: Remove age outlier rows where age > 100 (data entry error)
df_clean = df_clean[df_clean['age'] <= 100]

# Fill remaining missing age values with the median (robust to outliers)
df_clean['age'] = df_clean['age'].fillna(df_clean['age'].median())
print('Age after fix — missing:', df_clean['age'].isnull().sum(), '| max:', df_clean['age'].max())
```

```python
# ISSUE 2: loan_interest_rate — missing values and one negative value
print('Missing loan_interest_rate:', df_clean['loan_interest_rate'].isnull().sum())
print('Negative interest rates:', (df_clean['loan_interest_rate'] < 0).sum())
```

```python
# FIX ISSUE 2: Replace negative interest rate with NaN, then fill all missing with median
df_clean.loc[df_clean['loan_interest_rate'] < 0, 'loan_interest_rate'] = np.nan
df_clean['loan_interest_rate'] = df_clean['loan_interest_rate'].fillna(df_clean['loan_interest_rate'].median())
print('After fix — missing:', df_clean['loan_interest_rate'].isnull().sum(), '| min:', df_clean['loan_interest_rate'].min())
```

```python
# ISSUE 3: payment_default_on_file — 5 missing values
print('Missing payment_default_on_file:', df_clean['payment_default_on_file'].isnull().sum())
```

```python
# FIX ISSUE 3: Fill missing payment_default_on_file with mode (most frequent value)
df_clean['payment_default_on_file'] = df_clean['payment_default_on_file'].fillna(df_clean['payment_default_on_file'].mode()[0])
print('After fix — missing:', df_clean['payment_default_on_file'].isnull().sum())
```

```python
# ISSUE 4: emplyment_length — outlier values > 50 years (implausible)
print('Employment length > 50:', (df_clean['emplyment_length'] > 50).sum())
print('Max employment length:', df_clean['emplyment_length'].max())
```

```python
# FIX ISSUE 4: Remove rows where employment length > 50 (data entry errors)
df_clean = df_clean[df_clean['emplyment_length'] <= 50]
print('After fix — employment length max:', df_clean['emplyment_length'].max())
```

```python
# ISSUE 5: max_allowed_loan — 3 negative values (data entry/recording error)
print('Negative max_allowed_loan:', (df_clean['max_allowed_loan'] < 0).sum())
```

```python
# FIX ISSUE 5: Replace negative max_allowed_loan with NaN, then fill with median
df_clean.loc[df_clean['max_allowed_loan'] < 0, 'max_allowed_loan'] = np.nan
df_clean['max_allowed_loan'] = df_clean['max_allowed_loan'].fillna(df_clean['max_allowed_loan'].median())
print('After fix — negative count:', (df_clean['max_allowed_loan'] < 0).sum())
```

```python
# Verify: no remaining missing values in cleaned dataset
print('Final missing values:')
print(df_clean.isnull().sum())
print('\nFinal dataset shape:', df_clean.shape)
```

## Section 8: Encoding Categorical Variables
Code block leveraged and reused from: Seminar Session — Data Preparation Lab

```python
# Initialise LabelEncoder objects for each categorical column
le_home = LabelEncoder()
le_intent = LabelEncoder()
le_default = LabelEncoder()

# Encode home_ownership column into numerical values
df_clean['home_ownership_enc'] = le_home.fit_transform(df_clean['home_ownership'])

# Encode loan_intent column into numerical values
df_clean['loan_intent_enc'] = le_intent.fit_transform(df_clean['loan_intent'])

# Encode payment_default_on_file column into numerical values (N=0, Y=1)
df_clean['payment_default_enc'] = le_default.fit_transform(df_clean['payment_default_on_file'])

# Display encoding mappings for reference
print('Home Ownership encoding:', dict(zip(le_home.classes_, le_home.transform(le_home.classes_))))
print('Loan Intent encoding:', dict(zip(le_intent.classes_, le_intent.transform(le_intent.classes_))))
print('Payment Default encoding:', dict(zip(le_default.classes_, le_default.transform(le_default.classes_))))
```

## Section 9: Prepare Final Feature Sets
Code block leveraged and reused from: Seminar Session — Data Preparation Lab

```python
# Define features and target for CLASSIFICATION (Case Study A)
# max_allowed_loan is excluded as it is the regression target, not a classification feature
clf_features = ['age', 'income', 'home_ownership_enc', 'emplyment_length',
                'loan_intent_enc', 'loan_amount', 'loan_interest_rate',
                'loan_income_ratio', 'payment_default_enc', 'credit_history_length']

X_clf = df_clean[clf_features]
y_clf = df_clean['loan_approval_status']

print('Classification — Feature names:', X_clf.columns.tolist())
print('Classification — Data shape (X):', X_clf.shape)
print('Classification — Target shape (y):', y_clf.shape)
```

```python
# Define features and target for REGRESSION (Case Study B)
# Use only approved clients (loan_approval_status=0, i.e. Credit Application Acceptance = Yes)
# with positive max_allowed_loan values
reg_df = df_clean[(df_clean['loan_approval_status'] == 0) & (df_clean['max_allowed_loan'] > 0)]

reg_features = ['age', 'income', 'home_ownership_enc', 'emplyment_length',
                'loan_intent_enc', 'loan_amount', 'loan_interest_rate',
                'loan_income_ratio', 'payment_default_enc', 'credit_history_length']

X_reg = reg_df[reg_features]
y_reg = reg_df['max_allowed_loan']

print('Regression — Feature names:', X_reg.columns.tolist())
print('Regression — Data shape (X):', X_reg.shape)
print('Regression — Target shape (y):', y_reg.shape)
```
